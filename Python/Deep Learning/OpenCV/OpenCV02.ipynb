## 칼라이미지로 읽어서 gray 이미지로 변환하는 방법

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./images/bears.png")

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap="gray")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# 128 보다 작으면 검정색(0), 크면 흰색(255)로 변환
# 반환 : 기준값과 변환된 이미지
_, th_img = cv2.threshold(img_gray, 100, 255, cv2.THRESH_BINARY)

plt.imshow(th_img, cmap="gray")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# th : OTSU 로 결정된 threshold_value 값
th, th_img = cv2.threshold(img_gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

plt.imshow(th_img, cmap="gray")
plt.title(th)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
img2 = cv2.imread("./images/sudoku.png")
```

```python
img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap="gray")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
_, th_img = cv2.threshold(img_gray2, 80, 100, cv2.THRESH_BINARY)

plt.imshow(th_img, cmap="gray")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
block_size = 9
c = 5

th_img1 = cv2.adaptiveThreshold(img_gray2, 255,
                                  cv2.ADAPTIVE_THRESH_MEAN_C,
                                  cv2.THRESH_BINARY, block_size, c)
th_img2 = cv2.adaptiveThreshold(img_gray2, 255,
                                  cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                  cv2.THRESH_BINARY, block_size, c)   

plt.figure(figsize=(8,8))
plt.subplot(1, 2, 1)
plt.title("mean")
plt.imshow(th_img1, cmap="gray")
plt.xticks([])
plt.yticks([])

plt.subplot(1, 2, 2)
plt.title("gaussian")
plt.imshow(th_img2, cmap="gray")
plt.xticks([])
plt.yticks([])
```
