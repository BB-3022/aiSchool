```python
import requests
from bs4 import BeautifulSoup

# URL 주소 입력
url = 'https://emart.ssg.com/disp/category.ssg?dispCtgId=6000213114&sort=dcrt'

# requests.get() 메서드를 사용하여 해당 URL 의 웹 페이지 내용을 가져온다.
response = requests.get(url)

# 웹 페이지의 HTML 내용 파싱
soup = BeautifulSoup(response.content, 'html.parser')


# 할인상품
# class의 div 태그중 속성값이 'tmpl_itemlist'을 가지는 요소를 선택하여 할인 상품에 저장
할인상품 = soup.select_one('div.tmpl_itemlist')

#1. 상품명
# span 태그 중 class 값이 'mnemitem_goods_tit'에 해당하는 모든 정보를 상품명에 저장
상품명 = 할인상품.select('span.mnemitem_goods_tit')

# 2.정상가격
정상가격 = 할인상품.select('.old_price')

# 3.할인가격
할인가격 = 할인상품.select('.new_price')

# 4.할인율
할인 = 할인상품.select('.mnemitem_prd_per')

for idx in range(20):
    
    print(f"{idx + 1}번상품")
    
    print("상품 :", 상품명[idx].text)
    
    정상가격_element = 정상가격[idx].find('em', {'class': 'ssg_price'})
    print("정상가격 :", 정상가격_element.text)
    
    할인가격_element = 할인가격[idx].find('em', {'class': 'ssg_price'})
    print("할인가격 :", 할인가격_element.text)
    
    할인율_element = 할인[idx].find('span', {'class': 'blind'})
    할인율 = 할인율_element.find_next_sibling('span').text.strip()
    print("할인율 :", 할인율)
    
    print()
```

```python
# webdriver 위치 지정
driver = webdriver.Chrome()

# url 접속
url = "https://www.ssg.com/disp/category.ssg?dispCtgId=6000092893"
driver.get(url)
time.sleep(1)

# 스크롤을 X축 300, Y축 500 조절한다.
driver.execute_script("window.scrollTo(300, 600);")
time.sleep(1)

x_path = '//*[@id="area_sort"]/div[2]/div/ul/li[1]/div/div[1]'
menubar = driver.find_element(By.XPATH, x_path)
menubar.click()

x_path = '//*[@id="area_sort"]/div[2]/div/ul/li[1]/div/div[1]/div/ul/li[5]/a/span'
discount = driver.find_element(By.XPATH, x_path)
discount.click()

wait = WebDriverWait(driver, 30)

# 상세정보를 담을 리스트 생성
product_info_list = []

# 원래 페이지의 핸들 (현재 열려있는 탭) 가져오기
original_window_handle = driver.current_window_handle

# 각 상품을 클릭하고 상세페이지 정보를 가져올 반복문
for item_num in range(1, 2):  # 5개 상품에 대해 반복 (li 태그 5개에 대해)
    x_path = f'//*[@id="area_best"]/div[2]/div[4]/div/ul/li[{item_num}]/div/div[1]'
    
    # WebDriverWait를 사용하여 item 요소가 로드될 때까지 대기
    item = wait.until(EC.presence_of_element_located((By.XPATH, x_path)))
    
    product_link = item.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
    
    # 새로운 탭에서 링크를 열기
    driver.execute_script("window.open('', '_blank');")
    driver.switch_to.window(driver.window_handles[1])
    driver.get(product_link)

    # 상세페이지 정보를 담을 딕셔너리 생성
    product_info = {}

    # 상세페이지 정보 추출
    product_info['상품명'] = driver.find_element(By.CSS_SELECTOR, '.cdtl_info_tit_txt').text.strip()

    # 상세페이지 정보를 리스트에 추가
    product_info_list.append(product_info)

    # 이전 탭으로 돌아가기
    driver.close()
    driver.switch_to.window(original_window_handle)

# 작업이 끝나면 웹 드라이버 닫기
driver.quit()

# 상품 정보 리스트 출력
for product_info in product_info_list:
    print(product_info)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# webdriver 위치 지정
driver = webdriver.Chrome()

# url 접속
url = "https://www.ssg.com/disp/category.ssg?dispCtgId=6000092893&sort=dcrt"
driver.get(url)
time.sleep(1)

# 상세정보를 담을 리스트 생성
product_info_list = []

# 각 상품을 클릭하고 상세페이지 정보를 가져올 반복문
for item_num in range(1, 6):  # 5개 상품에 대해 반복 (li 태그 5개에 대해)
    x_path = f'//*[@id="ty_thmb_view"]/div[2]/div[4]/div/ul/li[{item_num}]/div/div[1]/div/a'
    item = wait.until(EC.element_to_be_clickable((By.XPATH, x_path)))
    product_link = item.get_attribute('href')
    
    # 새로운 탭으로 링크를 열기
    driver.execute_script("window.open('', '_blank');")
    driver.switch_to.window(driver.window_handles[1])
    driver.get(product_link)

    # 새로운 탭의 페이지가 완전히 로드될 때까지 대기
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.cdtl_info_tit_txt')))

    # 상세페이지 정보를 담을 딕셔너리 생성
    product_info = {}

    # 상세페이지 정보 추출
    product_info['상품명'] = driver.find_element(By.CSS_SELECTOR, '.cdtl_info_tit_txt').text.strip()
    product_info['원가'] = driver.find_element(By.CSS_SELECTOR, '.cdtl_optprice .ssg_price:nth-child(2)').text.strip()
    product_info['할인가격'] = driver.find_element(By.CSS_SELECTOR, '.cdtl_optprice .ssg_price:nth-child(1)').text.strip()

    # 원산지 정보 가져오기
    origin_element = driver.find_element(By.CSS_SELECTOR, '.cdtl_txt_origin .txt_origin')
    product_info['원산지'] = origin_element.text.strip()

    # 리뷰 정보 가져오기
    review_element = driver.find_element(By.CSS_SELECTOR, '.cdtl_grade_total')
    product_info['리뷰평점'] = review_element.text.strip()

    # 무이자 혜택 정보 가져오기
    installment_element = driver.find_element(By.CSS_SELECTOR, '.cdtl_card_benefit.v3 .cdtl_card_tit')
    product_info['무이자 혜택'] = installment_element.text.strip()

    # 상세페이지 정보를 리스트에 추가
    product_info_list.append(product_info)

    # 새로운 탭을 닫고 이전 페이지로 돌아가기
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(1)

# 작업이 끝나면 웹 드라이버 닫기
driver.quit()

# 상품 정보 리스트 출력
for product_info in product_info_list:
    print(product_info)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def get_product_details(driver, item_num):
    product_info = {}

    x_path = f'//*[@id="ty_thmb_view"]/div[2]/div[4]/div/ul/li[{item_num}]/div/div[1]/div/a'

    # 클릭 가능한 요소가 나타날 때까지 대기
    item = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, x_path)))
    product_link = item.get_attribute('href')

    # 새로운 탭에서 링크를 열기
    driver.execute_script("window.open('', '_blank');")
    driver.switch_to.window(driver.window_handles[1])
    driver.get(product_link)

    # 상세페이지 정보 추출
    product_info['상품명'] = driver.find_element(By.CSS_SELECTOR, '.cdtl_info_tit_txt').text.strip()
    product_info['원가'] = driver.find_element(By.CSS_SELECTOR, '.cdtl_optprice .ssg_price:nth-child(2)').text.strip()
    product_info['할인가격'] = driver.find_element(By.CSS_SELECTOR, '.cdtl_optprice .ssg_price:nth-child(1)').text.strip()

    # 원산지 정보 가져오기
    origin_element = driver.find_element(By.CSS_SELECTOR, '.cdtl_txt_origin .txt_origin')
    product_info['원산지'] = origin_element.text.strip()

    # 리뷰 정보 가져오기
    review_element = driver.find_element(By.CSS_SELECTOR, '.cdtl_grade_total')
    product_info['리뷰평점'] = review_element.text.strip()

    # 무이자 혜택 정보 가져오기
    installment_element = driver.find_element(By.CSS_SELECTOR, '.cdtl_card_benefit.v3 .cdtl_card_tit')
    product_info['무이자 혜택'] = installment_element.text.strip()

    # 새로운 탭을 닫고 이전 페이지로 돌아가기
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(1)

    return product_info

# webdriver 위치 지정
driver = webdriver.Chrome()

# url 접속
url = "https://www.ssg.com/disp/category.ssg?dispCtgId=6000092893&sort=dcrt"
driver.get(url)

# 상세정보를 담을 리스트 생성
product_info_list = []

# 각 상품을 클릭하고 상세페이지 정보를 가져올 반복문
for item_num in range(1, 6):  # 5개 상품에 대해 반복 (li 태그 5개에 대해)
    try:
        product_info = get_product_details(driver, item_num)
        product_info_list.append(product_info)
    except:
        pass

# 작업이 끝나면 웹 드라이버 닫기
driver.quit()

# 상품 정보 리스트 출력
for product_info in product_info_list:
    print(product_info)
```
