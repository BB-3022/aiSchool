```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

service_key = '16JJ62M2-16JJ-16JJ-16JJ-16JJ62M2IT'

url = f'http://safemap.go.kr/openApiService/data/getConvenienceStoreData.do?' \
    f'serviceKey={service_key}&numOfRows=20&pageNo=1&dataType=XML'

result = urlopen(url)
data = result.read()  # 데이터를 바이트 형태로 읽어옴

soup = BeautifulSoup(data, 'xml')  # 파싱할 때 'xml' 지정
items = soup.find_all('item')

for item in items:
    objt_id = item.find('OBJT_ID').text
    fclty_ty = item.find('FCLTY_TY').text
    fclty_nm = item.find('FCLTY_NM').text
    adres = item.find('ADRES').text
    ctprvn_cd = item.find('CTPRVN_CD').text
    sgg_cd = item.find('SGG_CD').text
    emd_cd = item.find('EMD_CD').text

    print(f'일련번호: {objt_id}')
    print(f'상권업종: {fclty_ty}')
    print(f'상호명: {fclty_nm}')
    print(f'지번주소: {adres}')
    print(f'시도코드: {ctprvn_cd}')
    print(f'시군구코드: {sgg_cd}')
    print(f'읍면동코드: {emd_cd}')
```
