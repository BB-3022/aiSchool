```python
import requests
from bs4 import BeautifulSoup

# URL 주소 입력
url = 'https://emart.ssg.com/disp/category.ssg?dispCtgId=6000213114&sort=dcrt'

# requests.get() 메서드를 사용하여 해당 URL 의 웹 페이지 내용을 가져온다.
response = requests.get(url)

# 웹 페이지의 HTML 내용 파싱
soup = BeautifulSoup(response.content, 'html.parser')


# 할인상품
# class의 div 태그중 속성값이 'tmpl_itemlist'을 가지는 요소를 선택하여 할인 상품에 저장
할인상품 = soup.select_one('div.tmpl_itemlist')

#1. 상품명
# span 태그 중 class 값이 'mnemitem_goods_tit'에 해당하는 모든 정보를 상품명에 저장
상품명 = 할인상품.select('span.mnemitem_goods_tit')

# 2.정상가격
정상가격 = 할인상품.select('.old_price')

# 3.할인가격
할인가격 = 할인상품.select('.new_price')

# 4.할인율
할인 = 할인상품.select('.mnemitem_prd_per')

for idx in range(20):
    
    print(f"{idx + 1}번상품")
    
    print("상품 :", 상품명[idx].text)
    
    정상가격_element = 정상가격[idx].find('em', {'class': 'ssg_price'})
    print("정상가격 :", 정상가격_element.text)
    
    할인가격_element = 할인가격[idx].find('em', {'class': 'ssg_price'})
    print("할인가격 :", 할인가격_element.text)
    
    할인율_element = 할인[idx].find('span', {'class': 'blind'})
    할인율 = 할인율_element.find_next_sibling('span').text.strip()
    print("할인율 :", 할인율)
    
    print()
```
