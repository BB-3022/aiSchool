```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

import time
import requests

# 크롬 드라이버로 접속
driver = webdriver.Chrome()

# driver를 이용해서 웹사이트 연결
#과일
url = 'https://front.homeplus.co.kr/list?categoryId=200010&categoryDepth=2'
# 채소
# url = 'https://front.homeplus.co.kr/list?categoryId=200029&categoryDepth=2'
# 수산물/건어물
# url = 'https://front.homeplus.co.kr/list?categoryId=200047&categoryDepth=2'
# 정육/계란
# url = 'https://front.homeplus.co.kr/list?categoryDepth=2&categoryId=200049'
# 델리/치킨/초밥
# url = 'https://front.homeplus.co.kr/list?categoryId=200118&categoryDepth=2'
# 우유/유제품
# url = 'https://front.homeplus.co.kr/list?categoryDepth=2&categoryId=200058'
# 생수/음료
# url = 'https://front.homeplus.co.kr/list?categoryDepth=2&categoryId=200100'
# 과자/시리얼
# url = 'https://front.homeplus.co.kr/list?categoryDepth=2&categoryId=200098'
# 라면/즉석식품/통조림
# url = 'https://front.homeplus.co.kr/list?categoryDepth=2&categoryId=200091'

# 브라우저 창 최대화
driver.maximize_window()

# driver.implicitly_wait(5)

# 웹 페이지 로딩
driver.get(url)

# response = requests.get(url)
# soup = BeautifulSoup(response.content, "html.parser")


# 원하는 요소에 접근하기 위한 대기 시간 설정(예: 10초)
wait = WebDriverWait(driver, 10)

# class="switch" 요소 클릭하여 닫기
switch_element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.switch')))
switch_element.click()

try:
    # 행사상품 체크박스 선택
    sale_element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="root"]/div/div[3]/div[2]/div[1]/div[3]/ul/li[1]/span/label/span[1]/i')))
    sale_element.click()
except Exception as e:
    print(e)   

#product_info_list = []


# 상품 크롤링 시작
product_info_list = []

# try:
# 목록 상품개수 확인      
items = driver.find_elements(By.CSS_SELECTOR, '.unitItemBox.mid.cardType.css-11rkkav-itemDisplayStyle')
num_products = len(items)
print("Number of products on the page:", num_products)

# 상품 url 크롤링
product_links = []
    
for item in items:
    link = item.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
    product_links.append(link)      
        
    #상품 담기
product_info = {}        
    # 목록의 url을 반복한다     
for link in product_links:
    driver.switch_to.window(driver.window_handles[-1])
    driver.get(link)
    time.sleep(2)  
#         try:
    product_info['구분'] = driver.find_element(By.CSS_SELECTOR, '.select').text.strip()
    product_info['상품명'] = driver.find_element(By.CSS_SELECTOR, '.prodTitle').text.strip()
    product_info['원가'] = driver.find_element(By.CSS_SELECTOR, '.price').text.strip()
    product_info['할인가격'] = driver.find_element(By.CSS_SELECTOR, '.priceValue').text.strip()

        # 오류남!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    link_element = driver.find_element(By.XPATH, '/html/head/link[3]')
    product_info['링크'] = link_element.get_attribute('href')
    product_info['이미지링크'] = driver.find_element(By.CSS_SELECTOR, 'img').get_attribute('src')

        # 추가!추가!
    product_info_list.append(product_info)

#         except:
#             pass

#         Close the current tab
#         driver.close()

#         driver.switch_to.window(driver.window_handles[0])
#         time.sleep(2) 

# finally:
#     Close the WebDriver
#     driver.quit()
# 출력
for index, product_info in enumerate(product_info_list, start=1):
    print("번호:", index)
    print("구분:", product_info['구분'])
    print("상품명:", product_info['상품명'])
    print("원가:", product_info['원가'])
    print("할인가격:", product_info['할인가격'])
    print("링크:", product_info['링크'])
    print("이미지링크:", product_info['이미지링크'])
    print("------------------------------")    
```
