# BeautifulSoup
- req를 통해서 받아온 텍스트 데이터를 컴퓨터가 이해할 수 있는 html 형태로 변환

```python
# 주의할점 : stone 이 아닌 일반 soup을 임포트 할 것!
# bs4 패키지의 BeautifulSoup 모듈을 임포트
from bs4 import BeautifulSoup as bs
```

```python
#2. 응답을 받은 데이터에서 뉴스 타이틀만 수집
```

```python
# 텍스트 형태의 파일을 lxml 형태로 변형시켜줘
soup = bs(res.text,"lxml")
```

```python
# html 태그에서 a 태그를 선택
# soup.select("a") -> 모든 a태그를 가지고 온다.
# . class / # id

#select("선택자") -> html 데이터에서 특정 요소만 수집
title=soup.select("a.news_tit")
```

```python
# 수집된 요소 중에 컨텐츠(텍스트)데이터만 추출
title[0].text
```

```python
for i in title :
    print(i.text)
```

# 네이버 금융정보를 수집

```python
res=req.get("https://finance.naver.com/")
```

```python
soup=bs(res.text,"lxml")
```

```python
# > : 자식, 공백 : 자손

# 팁! 부모자식을 의미하는 자식선택자를 잘 활용
# 수집하고자 하는 요소가 아이디, 클래스 구분자가 없다면 반드시 부모를 검사
# 부모태그가 구분자가 존재 할때까지 검사를 진행
# 구분자가 없는 경우에는 정확하지 않을 확률이 높음

title=soup.select("#_topItems1 > tr > th > a")
```

```python
title[0].text
```

```python
for i in title :
    print(i.text)
```

```python
# 선택자를 한번에 가지고 오는 방법
# 개발자도구에서 요소 우클릭 -> copy -> copy selector
# 주의점! 반드시 내가 선택한 하나만 가지고 온다.
# 복수개가 필요한 경우에는 선택자를 수정(ex nth-dhild())
# nth-child() : 가상클래스 선택자

soup.select("#_topItems1 > tr:nth-child(1) > th > a")
```
